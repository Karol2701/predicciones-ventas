```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename="/content/Sales_predictions_limpio"
df=pd.read_csv(filename)
df
```

```python
df2=df[["Outlet_Size","Item_Outlet_Sales"]].groupby(["Outlet_Size"]).mean().reset_index()
df2
```

**La mayor cantidad de ventas se concentra en las tiendas de tamaño mediano**

```python
df3=df[["Item_Type","Item_Outlet_Sales"]].groupby(["Item_Type"]).mean()
df3
```

```python
plt.figure(dpi=100)
plt.plot(df3.index,df3['Item_Outlet_Sales'].cumsum())
plt.ylabel('Nivel de ventas',fontsize=15)
plt.xlabel('Tipo de producto',fontsize=15)
plt.title('Ventas por tipo de producto',fontsize=15)
plt.legend(['Nivel de ventas','Tipo de producto'],bbox_to_anchor=(1.5,0.8))
plt.grid(axis='both')
##con el plt.xticks cambiamos la orientación como están escritas las palabras del eje X, si queremos cambiar las del eje y debemos utilizar plt.yticks
plt.xticks(rotation=90)
plt.show()
```
